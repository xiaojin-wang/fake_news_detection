```python
import pandas as pd
import numpy as np
import json
import os
import time
import tldextract
```

```python

def collect_tweets(sources, label):
    path = "{}/{}/".format(sources, label)
    news_dirs = os.listdir(path)
    tweet_dict_list = []
    cnt = 0
    for news_dir in news_dirs:
        if news_dir == '.DS_Store':
            continue
        if os.listdir(path + news_dir):
            tweets_dir = path + news_dir + '/tweets/'
            news_id = ''
            if sources == 'politifact':
                news_id = str(news_dir).replace('politifact','')
            elif sources == 'gossipcop':
                news_id = str(news_dir).split('-')[1]
            if news_id and os.path.exists(tweets_dir) and os.listdir(tweets_dir):
                tweets_files = os.listdir(tweets_dir)
                cnt += len(tweets_files)
                for tweet_file in tweets_files:
                    tweets_full_path = tweets_dir + tweet_file
                    with open(tweets_full_path, 'r') as f:
                        data = json.load(f)
                        fake = 1 if label == 'fake' else 0
                        tweet_dict = {'tweets_id':data['id'], 'created_at':data['created_at'], 'text':data['text'], 'news_id':news_id, 'user_id':data['user']['id'], 'location':data['user']['location'], 'verified':data['user']['verified'], 'source':data['source'],'followers_count':data['user']['followers_count'],'retweet_count':data['retweet_count'],'favorite_count':data['favorite_count'],'fake':fake}
                        tweet_dict_list.append(tweet_dict)
    print('tweet count',cnt)
    return pd.DataFrame(tweet_dict_list)
                
                
                    
                    
                
        
```

# gossipcop

```python
gossipcop_fake_df = collect_tweets('gossipcop','fake')
gossipcop_fake_df.head()
```

```python
gossipcop_fake_df.info()
```

```python
gossipcop_real_df = collect_tweets('gossipcop','real')
gossipcop_real_df.head()
```

```python
gossipcop_real_df.info()
```

```python
gossipcop = pd.concat([gossipcop_fake_df, gossipcop_real_df], axis = 0)
```

```python
gossipcop.to_csv('gossipcop_tweets.csv')
```
