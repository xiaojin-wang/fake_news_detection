Import necessary packages and read csv

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('gossipcop_with_users.csv')
```

```python
df = df.dropna()
```

```python
df_real = df[df.fake == 0]
```

DF which contains all the Real tweets.

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize = (30,20))
df_real['followers_count'].plot(ax=axes[0,0])
axes[0,0].set_title('followers_count')

df_real['friends_count'].plot(ax=axes[0,1])
axes[0,1].set_title('friends_count')

df_real['listed_count'].plot(ax=axes[1,0])
axes[1,0].set_title('listed_count')

df_real['favourites_count'].plot(ax=axes[1,1])
axes[1,1].set_title('favourites_count')

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2,figsize = (20,20))

df_real['geo_enabled'].value_counts().plot(ax=axes[0,0],kind = 'pie',autopct='%.1f')
axes[0,0].set_title('geo_enabled')

df_real['verified'].value_counts().plot(ax=axes[0,1],kind = 'pie',autopct='%.1f')
axes[0,1].set_title('verified')

df_real['has_extended_profile'].value_counts().plot(ax=axes[1,0],kind = 'pie',autopct='%.1f')
axes[1,0].set_title('has_extended_profile')

df_real['default_profile'].value_counts().plot(ax=axes[1,1],kind = 'pie',autopct='%.1f')
axes[1,1].set_title('default_profile')

plt.show()
```

```python
df_fake = df[df.fake == 1]
```
