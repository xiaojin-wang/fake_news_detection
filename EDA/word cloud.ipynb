```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
import nltk

import warnings
warnings.filterwarnings("ignore")
```

```python
pol_news = pd.read_csv('data/politifact_news.csv')
gos_news = pd.read_csv('data/gossipcop_news.csv')
```

```python
def tokenize(text):
    # tokenization (split text into tokens)
    # method: nltk.treebank
    treebank = nltk.tokenize.TreebankWordTokenizer()
    return treebank.tokenize(str(text))
```

```python
def remove_punctuation(text):
    # remove punctuations in text
    tokenizer = nltk.tokenize.RegexpTokenizer(r"\w+")
    text_without_punct = tokenizer.tokenize(' '.join(text))
    return text_without_punct
```

```python
nltk.download('stopwords')
def remove_stopwords(text):
    try:
        en_stopwords=nltk.corpus.stopwords.words('english')
        text_without_stopwords = [word for word in text if word not in en_stopwords]
        return ' '.join(word for word in text_without_stopwords)
    except:
        pass
```

```python
# all news in Politifact
texts = pol_news.text.apply(tokenize).apply(remove_punctuation).apply(remove_stopwords)
pol_text = " ".join(text for text in texts)
# Create and generate a word cloud image
wordcloud_pol_news = WordCloud(max_font_size=40, max_words=150, background_color="white").generate(pol_text)
# Display the generated image:
plt.imshow(wordcloud_pol_news, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# real news in Politifact
pol_real = pol_news[pol_news['fake'] == 0]
texts = pol_real.text.apply(tokenize).apply(remove_punctuation).apply(remove_stopwords)
pol_real_text = " ".join(text for text in texts)
# Create and generate a word cloud image
wordcloud_pol_real_news = WordCloud(max_font_size=40, max_words=150, background_color="white").generate(pol_real_text)
# Display the generated image:
plt.imshow(wordcloud_pol_real_news, interpolation='bilinear')
plt.axis("off")
plt.show()
```
